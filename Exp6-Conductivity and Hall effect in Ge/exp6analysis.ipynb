```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.style.use('science')
```

```python
data=np.loadtxt('ameros.txt')
x=data[:,0]
y=data[:,1]
plt.scatter(x, y, s=3)
plt.ylabel(r'$\sigma [1/Ohm.m]$')
plt.xlabel(r'$T^{-1} [1000/K]$')
plt.grid()
plt.yscale('log')
```

```python
def sigma(x, a, b,):
    return a*x+b
```

```python
popt, pcov =curve_fit(sigma, x, y)
print(popt)
print(np.sqrt(np.diag(pcov)))

range=np.linspace(np.min(x), np.max(x), 100)

plt.ylabel(r'$\sigma [1/Ohm.m]$')
plt.xlabel(r'$T^{-1} [1000/K]$')
plt.grid()
plt.semilogy(x, y, 'bo', markersize=2)

plt.plot(range, sigma(range, *popt), 'r--')
plt.legend(['Exp data', 'Fit function'], loc='upper right', prop={"size":9})
plt.savefig('sigma.png', dpi=500)
```
