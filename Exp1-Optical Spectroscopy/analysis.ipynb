```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import sklearn

plt.style.use('science')
```

```python
data1=np.loadtxt('1a.txt')
x1=data1[:,0]
y1=data1[:,1]
y_error=y1*0+0.002
x_error=x1*0
```

```python
plt.figure(figsize=(4,4))
plt.ylabel(r'Intensity (mV)')
plt.xlabel(r'Wavelength(nm)')
plt.grid()

plt.errorbar(x1, y1, y_error, x_error, 'bo', markersize=2, label='Exp data')
plt.legend(loc='upper left', prop={"size":9})
plt.savefig('1a.png', dpi=500)
```

```python
data2=np.loadtxt('meros_1b.txt')
x2=data2[:,0]
y2=data2[:,1]
y_error=y2*0+0.002
x_error=x2*0
```

```python

data_norm1 = y2/y1

print(data_norm1)
```

```python
plt.figure(figsize=(5,5))
plt.ylabel(r'Normalized Intensity (mV)')
plt.xlabel(r'Wavelength(nm)')
plt.grid()

plt.errorbar(x1, data_norm1, y_error, x_error, 'bo', markersize=2, label='Exp data')

plt.legend(loc='upper left', prop={"size":9})
plt.savefig('1b.png', dpi=500)
```

```python
data3=np.loadtxt('2.txt')
x3=data3[:,0]
y3=data3[:,1]
y_error=y3*0+0.002
x_error=x3*0
```

```python

data_norm2 = y3/y1
print(data_norm2)
```

```python
r=0.35
d=0.05
a=np.log(y1*(1-r)/y2)/d
v=3*10**8/(x2*10**(-9))
h=6.626*10**(-34)
e1=h*v*6.24*10**(18)
a2=np.sqrt(a)
```

```python

plt.figure(figsize=(4,4))
plt.ylabel(r'$\alpha (1/cm)$')
plt.xlabel(r'Energy hv(eV)')
plt.grid()
plt.scatter(e1, a, s=2, label='Exp data')
plt.legend(loc='upper left')
plt.semilogy()
plt.savefig('a1.png', dpi=500)
```

```python
plt.figure(figsize=(4,4))
plt.ylabel(r'$\alpha^{1/2} (1/cm^2)$')
plt.xlabel(r'Energy hv(eV)')
plt.grid()
plt.scatter(e1, a2, s=2, label='Exp data')
plt.legend(loc='upper left')

plt.savefig('a2.png', dpi=500)
```

```python
xm=np.where(e1<1.35)[0][0]
print(xm)
xy=np.where(e1<1.4)[0][0]
print(xy)
```

```python
def linear1 (x, a, b):
        return a *x + b

def linear2 (x, a, b):
        return a *x + b
```

```python
popt1, pcov1 =curve_fit(linear1, e1[xm:200], a2[xm:200])
print(popt1)
print(np.sqrt(np.diag(pcov1)))
plt.figure(figsize=(4,4))
range1=np.linspace(np.min(e1[xm:200])-0.05, np.max(e1[xm:200]+0.05), 100)

popt2, pcov2 =curve_fit(linear2, e1[18:33], a2[18:33])
print(popt2)
print(np.sqrt(np.diag(pcov2)))
plt.figure(figsize=(4,4))
range2=np.linspace(np.min(e1[18:33]), np.max(e1[18:33]), 100)

plt.ylabel(r'$\alpha^{1/2} (1/cm^2)$')
plt.xlabel(r'Energy hv(eV)')
plt.grid()
plt.scatter(e1, a2, s=2, label='Exp data')

plt.plot(range1, linear1(range1, *popt1), 'r--', label='Fit func. in order to find $E_g-E_p$')
plt.plot(range2, linear2(range2, *popt2), 'g--', label='Fit func. in order to find $E_g+E_p$')

plt.legend(loc='upper left', prop={"size":9})
plt.savefig('fit.png', dpi=500)
```
