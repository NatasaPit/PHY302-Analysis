```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.style.use('science')
```

```python
data=np.loadtxt('1_meros.txt')
x=data[:,0]
y=data[:,1]
plt.scatter(x, y, s=10)
plt.ylabel(r'$Illuminance(Watt/m^2)$')
plt.xlabel(r'Photodetector-Source distance (cm)')
plt.grid()
np.interp(60, x, y)
plt.savefig('1meros.png', dpi=500)
```

```python
data=np.loadtxt('3_meros_poly.txt')
x=data[:,0]
y1=data[:,1]
y2=data[:,2]

fig,ax = plt.subplots()
# make a plot
ax.plot(x, y1, color="red", marker="o", markersize=4)
# set x-axis label
ax.set_xlabel(r'Illuminance $(Watt/m^2)$', fontsize=14)
# set y-axis label
ax.set_ylabel(r"$I_s(mA)$",color="red",fontsize=14)
ax.grid()
ax2=ax.twinx()

# make a plot with different y-axis using second axis object
ax2.plot(x, y2,color="blue",marker="o", markersize=4)
ax2.set_ylabel(r"$U_o(V)$", color="blue",fontsize=14)
plt.ylim([3.5,9])
plt.title(r'Polycrystalline silicon photocell')

fig.savefig('3meros_poly.png', format='png', dpi=500, bbox_inches='tight')
```

```python
data1=np.loadtxt('1_meros.txt')
x=data1[:,0]
y=data1[:,1]
def sq (x, a, b,):
    return a/x**2+b
def linear (x, a, b):
    return a*x + b
```

```python
popt, pcov = curve_fit(sq, x, y)
print(popt)
print(np.sqrt(np.diag(pcov)))
```

```python
plt.scatter(x, y, s=10)

xaxis=np.linspace(np.min(x), np.max(x), 1000)
plt.plot(xaxis, sq(xaxis, *popt), '--', color='red')

plt.legend(['Exp data', 'Fit function'], loc='upper right', prop={"size":8})
plt.ylabel(r'$Illuminance(Watt/m^2)$')
plt.xlabel(r'Photodetector-Source distance (cm)')
plt.grid()
plt.savefig('1merosfit.png', dpi=500)
```
