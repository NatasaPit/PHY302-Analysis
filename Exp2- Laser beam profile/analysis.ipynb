```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.style.use('science')
```

```python
data9=np.loadtxt('9.txt')
x=data9[:,0]
y=data9[:,1]
def gauss (x, a, b, c):
    return a*np.exp(-(x-b)**2/(2*c**2))
def linear (x, a, b):
    return a*x + b
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(gauss, x, y, p0=(1,6.5, 5))
print(popt)
print(np.sqrt(np.diag(pcov)))
fwhm9=np.abs(popt[2])*2*np.sqrt(2*np.log(2))
print(fwhm9)
```

```python
plt.figure(figsize=(4,4))
plt.scatter(x, y, s=4)
xaxis=np.linspace(np.min(x), np.max(x), 1000)
plt.plot(xaxis, gauss(xaxis, *popt), '--', color='red')
plt.grid()
plt.legend(['Exp data', 'Fit function'], loc='upper left', prop={"size":8})
plt.ylabel(r'Intensity[mV]')
plt.xlabel(r'$Distance [mm]$')

plt.savefig('9.png', dpi=500)
```
