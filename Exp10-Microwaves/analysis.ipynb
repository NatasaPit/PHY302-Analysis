```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.style.use('science')
```

```python
data=np.loadtxt('5_1.txt')
x=data[:,0]*np.pi/180
y=data[:,1]
y_error=y*0+0.02
x_error=x*0+0.02
```

```python
def meros5_1(x, a, b,):
    return a*np.cos(x)+b
```

```python
def meros5_11(x, c, d,):
    return c*np.cos(x)**2+d
```

```python
popt1, pcov1 =curve_fit(meros5_1, x, y)
print(popt1)
print(np.sqrt(np.diag(pcov1)))

range=np.linspace(np.min(x), np.max(x), 100)
popt2, pcov2 =curve_fit(meros5_11, x, y)
print(popt2)
print(np.sqrt(np.diag(pcov2)))

plt.ylabel(r'Voltage (mV)')
plt.xlabel(r'Incident angle(rad)')
plt.grid()


plt.errorbar(x, y, y_error, x_error, 'bo', markersize=4)
plt.plot(range, meros5_1(range, *popt1), 'r--')
plt.plot(range, meros5_11(range, *popt2), 'g--')

plt.legend(['Fit func(cosx)','Fit func($cos^2x$)' ,'Exp Data'], loc='upper center', prop={"size":9})
plt.savefig('5_1meros.png', dpi=500)
```

```python
from scipy.signal import savgol_filter
data=np.loadtxt('10_katheti.txt')
x=data[:,0]
y=data[:,1]
y_filtered = savgol_filter(y, 7, 1)
y_error=y*0+0.02
x_error=x*0+0.1
```

```python
plt.errorbar(x, y, y_error, x_error, 'bo', markersize=4)
plt.plot(x, y_filtered, 'b-')

plt.grid()
plt.legend(['Interpolated data', 'Exp data'], loc='upper right', prop={"size":10})
plt.title(r'Brewster angle graph for perpendicular polarization')
plt.ylabel(r'Voltage(mV)')
plt.xlabel(r'Incident angle($^o$)')
plt.savefig('10_katheti.png', dpi=500)
```

```python
data=np.loadtxt('5_1.txt')
x=data[:,0]
y=data[:,1]
y_error=y*0+0.02
x_error=x*0+0.02
```

```python
def meros5_11(x, c, d,):
    return c*np.cos(np.pi*x/180)**2+d
```

```python
popt2, pcov2 =curve_fit(meros5_11, x, y)
print(popt2)
print(np.sqrt(np.diag(pcov2)))
range=np.linspace(np.min(x), np.max(x), 100)

plt.ylabel(r'Voltage (mV)')
plt.xlabel(r'Incident angle($^o$)')
plt.grid()

t=np.linspace(0,180,1000)
m0=y[0]
plt.errorbar(x, y, y_error, x_error, 'bo', markersize=4)
plt.plot(t, m0*np.cos(np.pi*t/180))
plt.plot(range, meros5_11(range, *popt2), 'r--')

plt.legend(['cosx','Fit func($cos^2x$)' ,'Exp Data'], loc='lower left', prop={"size":9})
plt.savefig('5_1meros1.png', dpi=500)
```
