```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

plt.style.use('science')
```

```python
data=np.loadtxt('1_cu.txt')
x=data[:,0]
y=data[:,1]
```

```python
def cu_1(x, a, b,):
    return a*x+b
```

```python
popt, pcov =curve_fit(cu_1, x, y)
print(popt)
print(np.sqrt(np.diag(pcov)))

range=np.linspace(np.min(x), np.max(x), 100)

plt.ylabel(r'Hall Voltage $U_H (mV)$')
plt.xlabel(r'Magnetic field (mT)')
plt.grid()
plt.plot(x, y, 'bo', markersize=3)
plt.plot(range, cu_1(range, *popt), 'r--')
plt.legend(['Exp data', 'Fit function'], loc='upper right', prop={"size":10})
plt.savefig('cu_1.png', dpi=500)
```
