```python
import numpy as np
datay=np.loadtxt('yellow.txt')
x=datay[:,0]
y=datay[:,1]
def gauss (x, a, b, c):
    return a*np.exp(-(x-b)**2/(2*c**2))
def linear (x, a, b):
    return a*x + b
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(gauss, x, y, p0=(1,600, 5))
print(popt)
print(np.sqrt(np.diag(pcov)))
```

```python
import matplotlib.pyplot as plt
import numpy as np
datay=np.loadtxt('yellow.txt')
x=datay[:,0]
y=datay[:,1]
plt.plot(x, y, 'o-', color='Yellow', markersize=2)
xaxis=np.linspace(np.min(x), np.max(x), 1000)
plt.plot(xaxis, gauss(xaxis, *popt), '--', color='orange')
plt.grid()
plt.legend(['Exp data', 'Fit function'], loc='upper left', prop={"size":8})
plt.title(r'Spectrum of yellow LED')
plt.ylabel(r'Amplitude')
plt.xlabel(r'$Wavelength [nm]$')
plt.xlim([560,620])
plt.savefig('yellowspectrum.png', dpi=500)
```

```python
from scipy.signal import savgol_filter
databb1=np.loadtxt('blackbody.txt')
x=databb1[:,0]
y=databb1[:,1]
y_filtered = savgol_filter(y, 151, 1)
def gauss (x, a, b, c):
    return a*np.exp(-(x-b)**2/(2*c**2))
def linear (x, a, b):
    return a*x + b
```

```python
plt.plot(x,y,'b-', alpha=0.3)
plt.plot(x, y_filtered, 'b-')
popt, pcov = curve_fit(gauss, x, y, p0=(1, 680, 5))
print(popt)
print(np.sqrt(np.diag(pcov)))
plt.plot(x, gauss(x, *popt), 'r-')
plt.grid()
plt.legend(['Exp data', 'Interpolated data', 'Fit function'], loc='upper right', prop={"size":6})
plt.title(r'Spectrum of blackbody for higher voltage value')
plt.ylabel(r'Amplitude')
plt.xlabel(r'$Wavelength [nm]$')
plt.savefig('blackbody1spectrumint.png', dpi=500)
```

```python
def planck(x, a, ):
	return a*x
```

```python
pl_data = np.loadtxt('planckdata.txt')
x = pl_data[:,0]
y = pl_data[:,1]
popt, pcov = curve_fit(planck, x, y)

print(popt)
print(np.sqrt(np.diag(pcov)))

range = np.linspace(np.min(x), np.max(x), 100)

plt.title(r'$V_f=f(1/\lambda)$')
plt.xlabel(r'$1/\lambda [1/nm]$')
plt.ylabel(r'$V_f$ [V]')
plt.grid()
plt.plot(x, y, 'bo')
plt.plot(range, planck(range, *popt), 'r--')
plt.legend(['Exp. data', 'Fit function'], loc='upper left', prop={"size":9})
plt.savefig('planck.png', dpi=500)
```
